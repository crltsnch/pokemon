```python
import pandas as pd
df=pd.read_csv('pokemon.csv')
```

```python
df=df.dropna()
```

```python
df
```

```python
def tipos():
    dic = {}
    for i in range(len(df)):
        if df.iloc[i]['Type 1'] not in dic:
            dic[df.iloc[i]['Type 1']] = 1
        else:
            dic[df.iloc[i]['Type 1']] += 1
    
    return dic
tipos()
```

```python
def buscarPokemon_Nombre():
    nom=input("Introduzca el pokemon que quiere buscar: ")
    listNom=[]
    for i in range (0, len(df)):
        if nom in df.iloc[i]['Name']:
            listNom.append(df.iloc[i]['Name'])
    return listNom
```

```python
def debiles():
    l = buscarPokemon_Nombre()
    tipo = ''
    for i in range(len(df)):
        if l[0] == df.iloc[i]['Name']:
            tipo = df.iloc[i]['Type 1']
    
    listadebiles=[]
    for i in range(len(df)):
        if tipo == df.iloc[i]['Type 2']:
            listadebiles.append(df.iloc[i]['Name'])
    
    return listadebiles

debiles()
    
```
